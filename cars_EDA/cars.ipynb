```python
performance_kpis = df[['engine_displacement','no_cylinder','max_torque_nm','max_power_bhp']]
efficiency_kpis = df[['fuel_type','fuel_tank_capacity']]
satisfaction_kpis = df[['rating','reviews_count']]
market_kpis = df[['starting_price','ending_price','body_type']]
comfort_kpis = df[['seating_capacity','transmission_type']]
```

```python
print("Performnace KPI's:")
print(performance_kpis.describe())
```

```python
# WE CAN PLOT HISTOGRAM BY TWO MEATHODS:
# 1. by putting values on our own
# 2. by automating that task by using numpy.histogram_bin_edges

#plt.figure(figsize=(15, 10))
#for i, col in enumerate(performance_kpis.columns, 1):
#    plt.subplot(2, 2, i)
#    sns.histplot(performance_kpis[col], bins=20, kde=True)
#    plt.title(f'Histogram of {col}')
#plt.tight_layout()
#plt.show()

#NEW

plt.figure(figsize=(15, 10))
for i, col in enumerate(performance_kpis.columns, 1):
    # Calculate the bin edges using numpy.histogram_bin_edges
    bin_edges = np.histogram_bin_edges(performance_kpis[col], bins='auto')  # 'auto' can be replaced with other strategies
    
    #Different Binning Strategies:
    #You can change the bins parameter in np.histogram_bin_edges to try different binning strategies:
    #
    #'auto': Automatically selects the best bin width.
    #'fd': Freedman-Diaconis estimator, good for skewed data.
    #'sturges': Sturges' formula, suitable for normal distributions.
    #'doane': An improved version of Sturges' formula.
    #'scott': Scott's normal reference rule.

    # Create a subplot for each column
    plt.subplot(2, 2, i)
    sns.histplot(performance_kpis[col], bins=bin_edges, kde=True)  # Use the calculated bin edges
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

Question 1: Which car has the highest engine displacement?

```python
highest_engine_displacement_car = df.loc[df['engine_displacement'].idxmax()]
print(f"Car with highest engine displacement:\n{highest_engine_displacement_car}\n")
```

Question 2: Correlation between engine displacement and max power

```python
correlation = df[['engine_displacement', 'max_power_bhp']].corr().iloc[0, 1]
print(f"Correlation between engine displacement and max power: {correlation:.2f}")
data = df[['engine_displacement', 'max_power_bhp']]
sns.heatmap(data.corr())
```

Question 3 : Which car model has the highest customer rating?

```python
highest_rated_car = df.loc[df['rating'].idxmax()]
print(f"Car with highest rating:\n{highest_rated_car}\n")
```

Question 4 : What is the price range (starting and ending prices) for different body types?

```python
price_range = df.groupby('body_type')[['starting_price','ending_price']].agg(['min','max'])
price_range.columns = ['starting_price_min','starting_price_max','ending_price_min','ending_price_max']
price_range
```

Question 5 : Which body type has the most models within the dataset?

```python
models = df['body_type'].value_counts()
most_models_body_type = models.idxmax()
model_cnt = models.max()
print(f"Body Type with the most models in the dataset is : {most_models_body_type} ({model_cnt} models)")
```
